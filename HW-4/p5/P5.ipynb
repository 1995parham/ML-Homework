## spiral

```python
import pandas as pd

df = pd.read_csv('spiral.txt', sep='\t')
X = df.values[:, :2]
y = df.values[:, 2]
```

```python
print(X.shape)
```

Initiates the DBSCAN algorithm with parameters that indicates better results. These parameters are guessed with try and error.

```python
from dbscan import DBSCAN

db = DBSCAN(eps=1.5, min_samples=3)
db.fit(X)
```

Show points on scatter plot. There are 2 clusters and one cluster for noises.

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8, 4))

for cl in range(db.label + 1):
    cluster = X[db._labels == cl]
    ax.scatter(cluster[:,0], cluster[:,1], label=cl)
noise = X[db._labels == -1]
ax.scatter(noise[:,0], noise[:,1], label='noise')
ax.legend()
```

```python
from sklearn.metrics import v_measure_score

v_measure_score(y, db._labels)
```

## D31

```python
df = pd.read_csv('D31.txt', sep='\t')
X = df.values[:, :2]
y = df.values[:, 2]
```

Initiates the DBSCAN algorithm with parameters that indicates better results. These parameters are guessed with try and error.

```python
from dbscan import DBSCAN

db = DBSCAN(eps=0.5, min_samples=5)
db.fit(X)
```

Show points on scatter plot. There are 31 clusters and one cluster for noises.

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8, 6))

for cl in range(db.label + 1):
    cluster = X[db._labels == cl]
    ax.scatter(cluster[:,0], cluster[:,1], label=cl)
noise = X[db._labels == -1]
ax.scatter(noise[:,0], noise[:,1], label='noise')
ax.legend(bbox_to_anchor=(1, 1))
```

```python
from sklearn.metrics import v_measure_score

v_measure_score(y, db._labels)
```

DBSCAN works well on a dataset that all clusters are similar to each other because this algorithm uses a single configuration on all clusters, so if there is a big difference between clusters causes issues to the algorithm.
