```python
from k_means import KMeans

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
markers = ('s', 'x', 'o', '^', 'v')

# reads data from the given csv and loads it into ndarray
df = pd.read_csv('data_kmeans_1.csv', header=None)
X = df.values

# initiates k-means algorithm
k = 2
km = KMeans(k=k, n_iter=50)
km.fit(X)

# draw the points on the plot
for cl in range(k):
    plt.scatter(X[km._labels == cl,0], X[km._labels == cl,1], color=colors[cl], marker=markers[cl], label=f'k={cl}')
plt.scatter(km._centroids[:,0], km._centroids[:,1], color='black', label='centroid')

plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper left')
plt.show()


plt.plot(range(1, len(km.dbi) + 1), km.dbi, marker='o')
plt.xlabel('Epochs')
plt.ylabel('davies bouldin index')
plt.show()
```
