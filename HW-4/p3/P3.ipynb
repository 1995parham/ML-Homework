```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
markers = ('s', 'x', 'o', '^', 'v')
```

```python
import pandas as pd

# reads data from the given csv
df1 = pd.read_csv('data_kmeans_1.csv', header=None)
df4 = pd.read_csv('data_kmeans_4.csv', header=None)
```

```python
from k_means import KMeans

# loads data to ndarray
X = df1.values

# initiates k-means algorithm
k = 2
km = KMeans(k=k, n_iter=50)
km.fit(X)

# draw the points on the plot
for cl in range(k):
    plt.scatter(X[km._labels == cl,0], X[km._labels == cl,1], color=colors[cl], marker=markers[cl], label=f'k={cl}')
plt.scatter(km._centroids[:,0], km._centroids[:,1], color='black', label='centroid')

plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper left')
plt.show()

plt.plot(range(1, len(km.dbi) + 1), km.dbi, marker='o')
plt.xlabel('Epochs')
plt.ylabel('davies bouldin index')
plt.show()

plt.plot(range(1, len(km.sse) + 1), km.sse, marker='o')
plt.xlabel('Epochs')
plt.ylabel('sum of squared errors')
plt.show()
```

```python
from sklearn.cluster import DBSCAN

# loads data to ndarray
X = df1.values

# initiates DBSCAN algorithm
dbscan = DBSCAN(eps=1, min_samples=5).fit(X)

# draw the points on the plot
for cl in np.unique(dbscan.labels_):
    plt.scatter(X[dbscan.labels_ == cl,0], X[dbscan.labels_ == cl,1], color=colors[cl], marker=markers[cl], label=f'k={cl}')

plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper left')
plt.show()
```
