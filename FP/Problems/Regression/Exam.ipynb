```python
import pandas as pd
```

```python
df_train = pd.read_csv('day.csv')
df_train.describe()
```

```python
y_train = df_train['cnt'].values
X_train = df_train.loc[:, 'season':'windspeed'].values
```

```python
df_X_test = pd.read_csv('x_regression_day.csv', header=None)
df_X_test.describe()
```

```python
df_y_test = pd.read_csv('y_regression_day.csv', header=None)
df_y_test.describe()
```

```python
X_test = df_X_test.iloc[:, 2:-2].values
y_test = df_y_test.values
```

```python
from sklearn.preprocessing import MinMaxScaler

sc = MinMaxScaler()
sc.fit(X_train)
X_test = sc.transform(X_test)
X_train = sc.transform(X_train)
```

```python
from sklearn.neural_network import MLPRegressor

mlp = MLPRegressor(hidden_layer_sizes=(200,), max_iter=20000)
mlp.fit(X_train, y_train)
```

```python
print(f'Training accuracy {mlp.score(X_train, y_train)}')
print(f'Test accuracy {mlp.score(X_test, y_test)}')
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(mlp.predict(X_test), y_test)
```
