```python
import numpy as np
import pandas as pd
```

```python
df_train = pd.read_csv('hepatitis.data', header=None)
df_train = df_train.replace('?', np.nan)
df_train.describe(include='all')
```

```python
X_train = df_train.loc[:, 1:18].values
y_train = df_train.loc[:, 0].values
```

```python
from sklearn.impute import SimpleImputer

imp = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imp.fit(X_train)

X_train = imp.transform(X_train)
```

```python
df_X_test = pd.read_csv('x_classification.csv', header=None)
df_X_test = df_X_test.replace('?', np.nan)
df_X_test.describe(include='all')
```

```python
df_y_test = pd.read_csv('y_classification.csv', header=None)
df_y_test = df_y_test.replace('?', np.nan)
df_y_test.describe(include='all')
```

```python
X_test = df_X_test.loc[:, 1:18].values
y_test = df_y_test.loc[:, 0].values
```

```python
from sklearn.impute import SimpleImputer

imp = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imp.fit(X_test)

X_test = imp.transform(X_test)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)
X = scaler.transform(X_test)
X_train = scaler.transform(X_train)
```

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(max_depth=5)
clf.fit(X_train, y_train)
```

```python
print(f'accuracy on train dataset: {clf.score(X_train, y_train)}')
print(f'accuracy on test dataset: {clf.score(X_test, y_test)}')
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=150, random_state=42)
clf.fit(X_train, y_train)
```

```python
print(f'accuracy on train dataset: {clf.score(X_train, y_train)}')
print(f'accuracy on test dataset: {clf.score(X_test, y_test)}')
```
