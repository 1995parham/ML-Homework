First of all load digits dataset from the sklean package.

```python
from sklearn import datasets

digits = datasets.load_digits()
```

Then load `Logistic Regression` from the `sklearn` package and then train it with loaded dataset. Here we are using the default values.

```python
from sklearn import linear_model

clf = linear_model.LogisticRegression()
clf.fit(digits.data[:-25], digits.target[:-25])
```

Use last 25 data as an examples for the trained model.

```python
import matplotlib.pyplot as plt
import numpy as np

for data in digits.data[-25:]:
    pixels = data
    pixels = np.array(pixels, dtype='uint8')
    pixels = pixels.reshape((8, 8))

    plt.title('Label is {label}'.format(label=clf.predict([data])))
    plt.imshow(pixels, cmap='gray')
    plt.show()
```
