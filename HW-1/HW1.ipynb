In the following section we want to compare between multiple values of *alpha* for *n = 10* and *steps = 150*. The best value for *alpha* around 1. we demostrate hypothesis function for `[0.1, 1, 1.5]`.

As you may see the following diagram I think alpha = 1, n = 10 and steps = 150 is better than other settings for given dataset.

#### alpha = 0.1

```python
import numpy as np
import regression

_, ax2 = plt.subplots()
_, ax3 = plt.subplots()
h, mse = regression.regression(data={d[0]: d[1] for d in dataset.values}, n=10, alpha=0.1, steps=150, lmb=0)

x = [d[0] for d in dataset.values]
y = [d[1] for d in dataset.values]
yh = [np.polynomial.polynomial.polyval(x, h) for x in x]

ax2.plot(x, y, 'r--', x, yh, 'b--')

ax3.plot(range(150), mse)
```
